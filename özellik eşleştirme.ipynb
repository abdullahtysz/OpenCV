```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# ana grnty ie aktar

chos = cv2.imread("chocalates.jpg",0) 
```

```python
plt.figure()
plt.imshow(chos,cmap = "gray")
plt.axis("off")
plt.show();
```

```python
cho = cv2.imread("toblerone.jpg")
```

```python
plt.figure()
plt.imshow(cho,cmap = "gray")
plt.axis("off")
plt.show();
```

```python
# orb tanmlaycs  : ana grnt ile aradmz grnt arasndaki noktalar tespit edecek
# ke , kenar vs gibi nesneye ait zellikler
```

```python
orb = cv2.ORB_create()
```

```python
# find the keypoints and descriptors with SIFT
# anahtar nokta tespiti  

kp1, des1 = orb.detectAndCompute(cho,None)
kp2, des2 = orb.detectAndCompute(chos,None)
```

```python
# eleme : BFmatcher ile 
# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True) # yntem norm hamming
 
# noktalar eletir
matches = bf.match(des1,des2)

# mesafeye gre srala
matches = sorted(matches, key = lambda x:x.distance)



# ilk 20 eleen resmi grntle
img_matches= cv2.drawMatches(cho,kp1,chos,kp2,matches[:10], None,flags=2) 
# fotograf pikselleri bydnde beklenlen sonucu vermeyebilir opencv ve 
#  TypeError: Required argument 'outImg' (pos 6) not found hatas verebilr
# bu durumda  None parametresi ile kty grmezden gelinebilir

plt.figure(figsize=(12,8))
plt.imshow(img_matches)
plt.axis("off")
```
