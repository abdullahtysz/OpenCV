```python
import keras
import numpy as np
import tensorflow as tf 
from  keras.preprocessing.image import load_img,img_to_array,ImageDataGenerator
```

```python
img = load_img("hakkinen.jpg")
type(img)
```

```python
# ilk ilem verinin tipini deitirmek

img_array = img_to_array(img)
type(img_array)
```

```python
# reshape 
reshaped_img_array = img_array.reshape((1,)+img_array.shape) # (1,100,100,3) gibi formata dntrmek iin
print("Reshaped:" , reshaped_img_array.shape)
```

```python
# data generator 
data_generator = ImageDataGenerator(rotation_range= 60,
                                   width_shift_range= 0.2,
                                   height_shift_range=0.2,
                                   zoom_range= 0.2,
                                   horizontal_flip= True,
                                   fill_mode="nearest") 
```

```python
counter = 0
for i in  data_generator.flow(reshaped_img_array, batch_size = 1 ,
                                    save_to_dir = "Desktop/aug",
                                    save_prefix = "abc",
                                    save_format = "jpg"): 
    counter +=1
    if counter == 75:
        break
```
