```python
import matplotlib.pyplot as plt
import numpy as np

import keras_ocr
```

```python

# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# Get a set of three example images
images = [keras_ocr.tools.read(img) for img in ["test1.jpg" ,"test2.jpg","/content/starbucks.jpg"]]
```

```python
images 
```

```python
# hafızada cok yer yok ise çökebilir . o yüzden colab gpu oneriliri
# Each list of predictions in prediction_groups is a list of
# (word, box) tuples.
predictions = pipeline.recognize(images)
```

```python


# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(18,18))
for ax, image, prediction in zip(axs, images, predictions): #zip ile sonrasdan içindedaha rahat gezmek için
    keras_ocr.tools.drawAnnotations(image=image, predictions=prediction, ax=ax)
```
