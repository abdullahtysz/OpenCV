## Binarizasyon

 Binarizasyon ile başlayalım. Binarizasyon, bir görüntüyü gri tonlamaya dönüştürme işlemidir. Pratik düzeyde gri tonlamalı görüntülerde, bir tür siyah ve beyaz, yalnızca gri tonlarından oluşur. Binarizasyon da Tesseract OCR motoru tarafından dahili olarak yapılır ancak ışık koşulları görüntüde tekdüze değilse, bu yöntem iyi bir çıktı vermemektedir.

## Rescaleing

 Rescaleing ,  görüntüyü küçültmek olan görüntüyü yeniden boyutlandırmak anlamına gelen Yeniden Ölçeklendirmeye sahibiz.

INTER_AREA kullanarak veya daha küçük karakterleri tanıyabilmemiz için görüntüyü daha büyük bir boyuta ölçekleyerek yapılır.

Bu, görüntü kalitesi açısından iyi olan INTER_CUBIC kullanılarak yapılabilir ancak yavaştır .

veya size iyi bir görüntü kalitesi vermeyebilecek olan  INTER_LINEAR'ı  kullanabiliriz , daha hızlı performans sağlar.

Neden yeniden boyutlandırmamız gerekiyor?

Tesseract en iyi DPI değeri en az 300 dpi olan görüntülerde çalışır, bu nedenle faydalı olabilir.

Ayrıca, bu günlerde geliştirilmiş kamera kalitesi ile 16K*16K piksellik görüntüler alıyoruz.

bu çok büyük boyuttur  ve Tesseract'ın bu tür görüntüleri işlemesi çok zaman alacaktır.

Görüntünün boyutunu kabul edilebilir sınıra değiştirmek bizim için daha mantıklıdır ve bu bizim için işe yarar.

Ayrıca, en boy oranının korunduğundan emin olmamız gerekiyor ve bunu hem ağırlık hem de yükseklik olduğunda yapabiliriz.
