```python
# Image Conversion from BGR to Grayscale using OpenCV

import cv2
image = cv2.imread('invoice.jpg') # Read the image 

cv2.imshow("image1.jpg", image)
cv2.waitKey(0)                   # image will not show until this is called
cv2.destroyWindow('image1.jpg')  # make sure window closes cleanly

# The imported image is in BGR format 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imwrite("test4.jpg", gray)

cv2.imshow("image2.jpg", gray)
cv2.waitKey(0)                   # image will not show until this is called
cv2.destroyWindow('image2.jpg')  # make sure window closes cleanly
```
