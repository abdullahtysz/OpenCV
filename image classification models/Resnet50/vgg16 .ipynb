```python
from tensorflow.keras.applications.vgg16 import VGG16
from keras.preprocessing import image # görüntü işlemek için
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
import numpy as np 
```

```python
model = VGG16(weights="imagenet")
```

```python
test_path = "elephant.png"
# keras ile imageımı okuyabilirim 
img = image.load_img(test_path, target_size = (224,224)) # yenıden boyutlandırdım
img = image.img_to_array(img)   # arraye cevirdim
img = np.expand_dims(img, axis = 0 ) # 4 boyutlu veriye ceviriyorum . 0. indexe uyguluyorum
img = preprocess_input(img)  

prediction = model.predict(img)
prediction 
```

```python
prediction  = decode_predictions(prediction, top = 3 )[0] # isteğim formata dönüştürdüm ve tahminlerimi gösterdi 
prediction
```
