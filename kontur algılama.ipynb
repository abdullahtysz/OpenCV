```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# resmi içeri aktar
```

```python
img = cv2.imread("kontur.png",0) # sb aldım
```

```python
plt.figure(), plt.imshow(img,cmap = 'gray'), plt.axis("off");
```

 findContours() görüntü rengindeki değişikliği algılar ve bunu kontur olarak işaretler.

```python
# konturları tespit ediyorum
# konturların cıktısı contours, hierarch(dış ve iç conturları görmek için)


contours, hierarch =  cv2.findContours(img , mode= cv2.RETR_CCOMP, method=cv2.CHAIN_APPROX_SIMPLE) 

# yatay , dikey ve çapraz olarak bölümleri sıkıştırıyor ve sadece uc noktaları bırakıyor
```

```python
# matris oluşturuyorum 

external_contour = np.zeros(img.shape)  # dış konturlarım için
internal_contour = np.zeros(img.shape)  # iç konturlarım için 
```

```python
for i in range(len(contours)):
    
    # external
    if hierarch [0][i][3] == -1 : 
        cv2.drawContours(external_contour,contours, i , 255,-1) # 255, renk , -1 kalınlık 
    else : # internal
        cv2.drawContours(internal_contour,contours, i , 255,-1)
```

```python
plt.figure() 
plt.imshow(external_contour, cmap = "gray")
plt.axis("off");
```

```python
plt.figure() 
plt.imshow(internal_contour, cmap = "gray")
plt.axis("off");
```
