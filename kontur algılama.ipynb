```python
img = cv2.imread("kontur.png",0) # sb aldm
```

```python
plt.figure(), plt.imshow(img,cmap = 'gray'), plt.axis("off");
```

 findContours() grnt rengindeki deiiklii alglar ve bunu kontur olarak iaretler.

```python
# konturlar tespit ediyorum
# konturlarn ckts contours, hierarch(d ve i conturlar grmek iin)


contours, hierarch =  cv2.findContours(img , mode= cv2.RETR_CCOMP, method=cv2.CHAIN_APPROX_SIMPLE) 

# yatay , dikey ve apraz olarak blmleri sktryor ve sadece uc noktalar brakyor
```

```python
# matris oluturuyorum 

external_contour = np.zeros(img.shape)  # d konturlarm iin
internal_contour = np.zeros(img.shape)  # i konturlarm iin 
```

```python
for i in range(len(contours)):
    
    # external
    if hierarch [0][i][3] == -1 : 
        cv2.drawContours(external_contour,contours, i , 255,-1) # 255, renk , -1 kalnlk 
    else : # internal
        cv2.drawContours(internal_contour,contours, i , 255,-1)
```

```python
plt.figure() 
plt.imshow(external_contour, cmap = "gray")
plt.axis("off");
```

```python
plt.figure() 
plt.imshow(internal_contour, cmap = "gray")
plt.axis("off");
```
