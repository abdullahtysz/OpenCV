```python
import cv2 
import os
```

                             A  A M A L A R

```python
path = "images"  # kendi kameram ile kaydettiim grntleri burada toplayacam
```

```python
# resim boyutu
imgWidth = 180
imgHeight = 128 
```

```python
# video yakalama
cap = cv2.VideoCapture(0) # dahili kameram
cap.set(3,640)   # genilik 
cap.set(4,480)   # boyut
cap.set(10,180)  # aydnlk oranlar
```

```python
# klasr oluturma ( yakaladm grntleri toplayacam)


global countFolder
def saveDataFunc():
    global countFolder
    countFolder = 0  # en bata herhang br folderm yok 
    while os.path.exists(path + str(countFolder)) : # eer 0 var ise 1 diye devam et klasrlemeye
        countFolder += 1
    os.makedirs(path+str(countFolder)) 

    
saveDataFunc()    



# kameramamdan veri toplamaya balayacagm 


count = 0
countSave = 0

while True :
    success, img = cap.read()
    
    if success: # eer kameram alyorsa 
        img = cv2.resize(img, (imgWidth, imgHeight))  # resimlerimin boyutlarn klttm 
        
        #ok hzl veri depolayaca iin ayn samplelar art arda depolamasna gerek yok 
        #bu yzden 5 fotografta 1 ini depoluyorum
        if count % 5 == 0 :
            cv2.imwrite(path+str(countFolder)+ "/"+str(countSave)+"_"+".png",img)
                                # nereye koyacam belirledim 
            countSave +=1 # her bir resim kaydedildiinde  1 artsn
            print(countSave)
        count += 1
        
        cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()        
        # artk veri toplayabilirim
```
