```python
import cv2 
import os
```

                             A Ş A M A L A R

```python
path = "images"  # kendi kameram ile kaydettiğim görüntüleri burada toplayacağım
```

```python
# resim boyutu
imgWidth = 180
imgHeight = 128 
```

```python
# video yakalama
cap = cv2.VideoCapture(0) # dahili kameram
cap.set(3,640)   # genişlik 
cap.set(4,480)   # boyut
cap.set(10,180)  # aydınlık oranları
```

```python
# klasör oluşturma ( yakaladığım görüntüleri toplayacağım)


global countFolder
def saveDataFunc():
    global countFolder
    countFolder = 0  # en başta herhangı bır folderım yok 
    while os.path.exists(path + str(countFolder)) : # eğer 0 var ise 1 diye devam et klasörlemeye
        countFolder += 1
    os.makedirs(path+str(countFolder)) 

    
saveDataFunc()    



# kameramamdan veri toplamaya başlayacagım 


count = 0
countSave = 0

while True :
    success, img = cap.read()
    
    if success: # eğer kameram çalışıyorsa 
        img = cv2.resize(img, (imgWidth, imgHeight))  # resimlerimin boyutlarını küçülttüm 
        
        #çok hızlı veri depolayacağı için aynı sampleları art arda depolamasına gerek yok 
        #bu yüzden 5 fotografta 1 ini depoluyorum
        if count % 5 == 0 :
            cv2.imwrite(path+str(countFolder)+ "/"+str(countSave)+"_"+".png",img)
                                # nereye koyacağımı belirledim 
            countSave +=1 # her bir resim kaydedildiğinde  1 artsın
            print(countSave)
        count += 1
        
        cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()        
        # artık veri toplayabilirim
```
